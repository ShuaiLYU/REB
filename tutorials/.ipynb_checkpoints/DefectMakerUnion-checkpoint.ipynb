```python
from DMlib.defect_maker_base import DefectMakerBase,gen_random_noise_img
from DMlib.cut_out import CutOut,CutOutScar
from DMlib.cut_paste import CutPaste,CutPasteScar
from DMlib.bezier_mask import BezierMaskGenerator
from torchvision import transforms
import torch
from  functools import  partial
import numpy as np
def merge_dict(src,supplement):
    for k,v in supplement.items():
        assert(k  not in src.keys())
        src[k]=v
    return src

defect_makers=["CutPaste","CutPasteScar","CutOut","CutOutScar"]
from collections import OrderedDict
class DefectMakers(DefectMakerBase):
    
    def __init__(self,num_labels,maker_params,dataset,transform_before,transform_after,**kwargs):
        DefectMakerBase.__init__(self,dataset,transform_before,transform_after,**kwargs)

        self.set_labels(num_labels,maker_params)
        self.makers=[]
        for idx,maker_param in enumerate(maker_params):
            assert(maker_param.name in defect_makers),"{}".format(maker_param.name)
            maker=(eval(maker_param.name)(dataset=dataset,**maker_param))
            key=maker_param.name+str(idx+1)
            self.makers.append((key,maker))
            
        self.kwargs=kwargs
    def set_labels(self,num_labels,maker_params):
        self.num_labels=num_labels
        self.labels=[]
        for maker_param in maker_params:
            class_name=maker_param.name
            label=np.zeros(shape=(num_labels,),dtype=np.int32)
            label[0]=1
            if "Scar" in class_name: label[1]=1
            if "CutPaste" in class_name: label[2]=1
            if "CutOut" in class_name: label[3]=1
            if maker_param.get("prob_bezier_mask",0)==1:label[4]=1
            self.labels.append(label)
       
    
    def make_defect(self, sample):
        # print(sample)
        defect_sample={}
        for idx,(key,maker) in enumerate(self.makers):
            # print(maker.prob_bezier_mask)
            instance=maker.get_defect_instance(sample)
            dft_img=maker.run_fuse(sample["img"],instance)
            label=self.labels[idx]
            data= {key+"_ins":instance, key+"_img":dft_img,key+"_label":label}
            merge_dict(defect_sample,data)
        return defect_sample
    
    def __getitem__(self, idx):
        sample = self.dataset[idx]
        # print(sample)
        assert ( isinstance(sample,dict))
        sample["label"]=np.zeros(shape=(self.num_labels,),dtype=np.int32)
        #1. get img and perform img trans
        if  self.transform_before is not None:
            sample["img"]=self.transform_before(sample["img"])
        #2. make defect on img
        dft_sample= self.make_defect(sample)
        

        #merge
        data = {k: v for d in [sample, dft_sample] for k, v in d.items()}
        #3. perform trans on defected_img
        if  self.transform_after is not None:
            data= {k:self.transform_after(v) if "img" in k else v for k, v in data.items() }
        return  data
    def get_collate_fn(self):

        def cp_collate(items,key_label_map,label_style,mask_stride):
            xs,ys,masks,reconsts=[],[],[],[]
            for item in items:
                for k,val in key_label_map.items():
                    k=k if k=="img" else k+"_img"
                    item[k]=np.array(item[k])
                    xs.append(torch.tensor(item[k]) if not torch.is_tensor(item[k]) else item[k] ) 
                    ys.append(torch.tensor(val) if not torch.is_tensor(val) else val) 
                    if "mask" in label_style:
                        if k =="img": 
                            shape=[ int(s/mask_stride) for s in xs[-1].shape[1:] ]
                            mask =torch.zeros(shape)
                        else:
                            mask=np.array(item[k.replace("img","ins")].get_imshow_mask(fg=val,factor=1/mask_stride))
                        mask=(torch.tensor(mask) if not torch.is_tensor(mask) else mask)
                        masks.append(mask)    
                    if "reconst" in label_style:
                        reconsts.append(torch.tensor(item["src_img"]) if not torch.is_tensor(item["src_img"]) else item["src_img"] ) 
            xs,ys=torch.stack(xs),torch.stack(ys)
            if len(masks)>0:masks=torch.stack(masks)
            if len(reconsts)>0:reconsts=torch.stack(reconsts)
            if label_style =="class":
                return xs,ys
            elif label_style=="mask":
                return xs,masks
            elif label_style=="class_mask":
                return xs,[ys,masks]
            elif label_style=="reconst":
                reconsts=F.interpolate(reconsts,scale_factor=1/mask_stride,mode='bilinear').permute(0,2,3,1)
                return xs,reconsts
            elif label_style=="class_reconst":
                reconsts=F.interpolate(reconsts,scale_factor=1/mask_stride,mode='bilinear').permute(0,2,3,1)
                # print(xs.shape,reconst.shape)
                return xs,[ys,reconsts]
            elif label_style=="class_reconst_mask":
                reconsts=F.interpolate(reconsts,scale_factor=1/mask_stride,mode='bilinear').permute(0,2,3,1)
                # print(xs.shape,reconst.shape)
                return xs,[ys,reconsts,masks]
            else:
                raise Exception("got an unexpected param: label_style") 
                        
        
        label_style=self.kwargs.get("label_style","class")
        if label_style=="both":label_style="class_mask"
        assert(label_style in ["class","mask","class_mask","reconst","class_reconst","class_reconst_mask"])
        mask_stride=self.kwargs.get("mask_stride",1)
    
        
        key_label_map=OrderedDict()
        key_label_map["img"]=np.zeros(shape=(self.num_labels,),dtype=np.int32)
        for idx,(key,maker) in enumerate(self.makers):
            key_label_map[key]=self.labels[idx]
    
        return partial(cp_collate,key_label_map=key_label_map,label_style=label_style,mask_stride=mask_stride)
```
