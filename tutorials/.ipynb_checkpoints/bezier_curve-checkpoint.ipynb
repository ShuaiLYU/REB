```python

from PIL import Image
import matplotlib.pyplot as plt

patch_img=Image.fromarray(mask.astype(np.uint8)).convert("RGB")
     
    
mask_img=  Image.new("L",patch_img.size,255)
# mask_img =  Image.fromarray(np.full((patch_img.size[1],patch_img.size[0]),255,np.uint8))

    
    
rot_deg=10


patch_r = patch_img.rotate(rot_deg, expand=True)  # rotate

mask_r=mask_img.rotate(rot_deg, expand=True)  # rotate



big_img=  Image.new("L",(100,100),255)


big_img1=big_img.copy()
big_img1.paste(patch_r.convert("L"),(0,0,patch_r.size[0],patch_r.size[1]),mask=mask_r)
# big_img1.paste(mask_r,(0,0,patch_r.size[0],patch_r.size[1]),None)


plt.figure(figsize=(20,20))
plt.subplot(161),plt.imshow(patch_img)
plt.subplot(162),plt.imshow(mask_img)
plt.subplot(163),plt.imshow(patch_r)
plt.subplot(164),plt.imshow(mask_r)
plt.subplot(165),plt.imshow(big_img)
plt.subplot(166),plt.imshow(big_img1)


        
```
